```python
import re #regular expression
import pandas as pd
```

```python
f = open ('WhatsApp Chat with MCKVians ka Karnama2 üë©‚Äçüéìüéìüë®‚Äçüéì.txt','r',encoding='utf-8') #file handling to open stream
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s' #regex101.com tool
```

```python
messages=re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
df = pd.DataFrame({'user message': messages, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M - ')
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
```

```python
df
```

```python
df.drop(columns=['user message'], inplace=True)
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df.drop(columns=['date'], inplace=True)
```

```python
df.rename(columns={'only_date': 'date'}, inplace=True)
```
